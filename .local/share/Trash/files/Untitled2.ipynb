```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame(columns=['Name','Percentage','Age'])
df.loc[0]=['Misbah',79,19]
df.loc[1]=['Sadika',95,20]
df.loc[2]=['Sadiya',70,20]
df.loc[3]=['Adeeba',80,21]
df.loc[4]=['Sara',94,21]
df.loc[5]=['Falak',80,20]
df.loc[6]=['Tehzeeb',87,21]
df.loc[7]=['Iqra',78,20]
df.loc[8]=['Hema',94,21]
df.loc[9]=['Ayesha',73,20]
df['remark']=None
df.drop(labels=['remark'],axis=1,inplace=True)
df.dropna(axis=0,inplace=True)
print(df)
df.plot(x="Name", y="Percentage")
plt.xlabel("Name VS Percentage")
plt.xlabel("Name of Students")
plt.ylabel("Percentage")
plt.show           
```
