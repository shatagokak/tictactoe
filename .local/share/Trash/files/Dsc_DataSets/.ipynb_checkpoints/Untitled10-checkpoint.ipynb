```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

iris=pd.read_csv("Iris.csv")

iris['Species'].value_counts().plot.bar()
plt.title('Iris Species %')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

iris=pd.read_csv("Iris.csv")

plt.pie(x=iris['Species'].value_counts())
plt.title('Iris Species %')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

iris=pd.read_csv("Iris.csv")

plt.hist(x=iris['Species'])
plt.title('Iris Species %')
plt.show()
```
