```python
import numpy as np
import matplotlib.pyplot as plt
subjects=['OT','CN IT','WT','DSc','JAVA','PYTHON','BLOCK CHAIN']
marks=[48,44,49,42,43,46,47]
plt.pie(marks,labels=subjects)
plt.xlabel('subjects')
plt.ylabel('marks')
plt.title('marks vs subjects pie chart')
plt.show()

plt.bar(subjects,marks)
plt.xlabel('subjects')
plt.ylabel('marks')
plt.title('marks vs subjects pie chart')
plt.show()
```
