```python
import numpy as np
import matplotlib.pyplot as plt

rand_arr=np.random.randint(0,100,50)

print("\n ****** Random numbers ******* \n",rand_arr)

x=rand_arr
y=x
plt.plot(x,color='red')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Line Chart")
plt.show()



plt.scatter(x,y,marker='s',c='blue')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Scatter")
plt.show()


plt.hist(x,color='green')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Histogram")
plt.show()


plt.boxplot(x)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Box Plot")
plt.show()
```
